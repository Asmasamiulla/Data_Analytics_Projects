```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv(r"C:\Users\Personal\Downloads\archive\netflix_titles.csv")
df.head(2)
```

### Visualisation

### pie chart

```python
df['type'].value_counts().plot(kind = 'pie',autopct = '%1.1f%%')
```

### line plot

```python
sns.lineplot(data=df, x="type", y="release_year")
plt.title('Line Plot of Release Year vs type')
plt.xlabel('type')
plt.ylabel('release_year')
plt.show()
```

### Bar Plot

```python
plt.figure(figsize=(15, 6))
df['rating'].value_counts(normalize=True).plot.bar()
plt.title('Distribution of Rating Categories')
plt.xlabel('Rating')
plt.ylabel('Relative Frequency')
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.scatterplot(x='release_year', y='rating', data=df)
plt.title('Scatterplot of Release Year vs Rating')
plt.xlabel('release_year')
plt.ylabel('rating')
plt.show()
```

the Scatterplote shows the relationship between the release year and ratings of movies in a dataset.

### CountPlot

```python
plt.figure(figsize=(15, 6))
sns.countplot(x='rating', data=df, hue="type")
plt.show()
```

count plot It's used to show the counts of observations in each category using bars.
